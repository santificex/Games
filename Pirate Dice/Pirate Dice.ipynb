```python
import numpy as np
from scipy import stats as sp
```

```python
print('Number of players')
p=int(input())
```

```python
print('Number of dice per player')
d=int(input())
```

```python
print('Number of simulations')
s=int(input())-1
```

```python
ronda=1
for x in range(0,p*d+1):
    uno=[]
    dos=[]
    tres=[]
    cuatro=[]
    cinco=[]
    seis=[]

    eins=[]
    zwei=[]
    drei=[]
    vier=[]
    funf=[]
    sechs=[]

    for i in range(0,s+1):
        j=np.random.randint(1,7,ronda)
        j=list(j)
        one=j.count(1)
        uno.append(one)
        two=j.count(2)
        dos.append(two)
        three=j.count(3)
        tres.append(three)
        four=j.count(4)
        cuatro.append(four)
        five=j.count(5)
        cinco.append(five)
        six=j.count(6)
        seis.append(six)

    for i in range(0,s):
        eins.append(uno[i])
        zwei.append(uno[i]+dos[i])
        drei.append(uno[i]+tres[i])
        vier.append(uno[i]+cuatro[i])
        funf.append(uno[i]+cinco[i])
        sechs.append(uno[i]+seis[i])
    
        modas=list([sp.mode(eins)[0],sp.mode(zwei)[0],sp.mode(drei)[0],sp.mode(vier)[0],sp.mode(funf)[0],sp.mode(sechs)[0]])
        maxmodas=max(modas)
        
    
    if ronda == 1:
        pass
    elif ronda > p*d:
        pass
    else:
        print('Con '+str(ronda)+' dados:')
        print('Num max de 1´s')
        print(sp.mode(eins)[0])
        print('Num max de otros numeros')
        print(maxmodas)
        print()
    ronda+=1
    
```
